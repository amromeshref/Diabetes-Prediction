## 1) Importing Packages

#### Add the repository directory path to the Python path

```python
import os
import sys

REPO_DIR_PATH = os.path.normpath(os.path.join(
    os.path.join(os.path.dirname(os.getcwd()))))

sys.path.append(REPO_DIR_PATH)
```

```python
from src.train import Trainer
import matplotlib.pyplot as plt
import torch
```

```python
trainer = Trainer()
```

### --------------------------------------------------------------------------------------------------------------------------------------------------------

## 2) Loading the Data

```python
X_train, y_train, X_test, y_test = trainer.load_data()
```

```python
print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_test:", y_test.shape)
```

### --------------------------------------------------------------------------------------------------------------------------------------------------------

## 3) Training the Model

```python
# Instantiate an ANN Model
model = trainer.instantiate_model()
```

```python
# Instantiate an optimizer for the model
learning_rate = 0.001
optimizer = trainer.instantiate_optimizer(model, learning_rate)
```

```python
EPOCHS = 500
```

```python
loses = trainer.train(X_train, y_train, model, optimizer, EPOCHS)
```

### Ploting the Loses vs Epochs

```python
plt.plot(range(EPOCHS),loses)
plt.ylabel('Loss')
plt.xlabel('Epoch')
```

### --------------------------------------------------------------------------------------------------------------------------------------------------------

## 4) Evaluating the Model

#### Evaluating on Training Data

```python
accuracy = trainer.evaluate(X_train, y_train, model)
print("Accuracy =", accuracy)
```

#### Evaluating on Testing Data

```python
accuracy = trainer.evaluate(X_test, y_test, model)
print("Accuracy =", accuracy)
```

### --------------------------------------------------------------------------------------------------------------------------------------------------------

## 5) Saving the Model

```python
torch.save(model, "../models/model.pt")
```
